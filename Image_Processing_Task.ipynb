```python
#imports
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
def show_image(image, grey=False):
    if grey:
        plt.imshow(image,cmap='gray', vmin = 0, vmax = 255)
    else:
        plt.imshow(image)

    plt.show()
```

# Align and fix the following Paragrapgh

```python
text_image = cv2.imread('badtext.jpg')

show_image(text_image)
```

```python
rows = text_image.shape[0]
cols = text_image.shape[1]
```

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),40,1) 

rotated_image = cv2.warpAffine(text_image,M,(cols,rows))
show_image(rotated_image)
```

```python
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
HMG = cv2.filter2D(rotated_image, -1, kernel)
show_image(HMG)
```

```python
kernel = np.ones((2,2), np.uint8)

img_erosion = cv2.erode(HMG, kernel, iterations=1)
show_image(img_erosion)
```
